```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Результаты_опросов.csv", header=None, names=["Дата", "Одобряю", "Не одобрять"])
```

```python
df
```

```python
df["Дата"] = pd.to_datetime(df["Дата"], format="%d.%m.%Y")
```

# **визуализация данных об утверждении при Борисе Ельцине**

```python
plt.figure(figsize=(18, 8))
plt.plot(df["Дата"], df["Одобряю"], label="Одобряю(%)", color="blue", marker="o")
plt.title("Процентное изменение поддержки Бориса Ельцина (1994-2000)(Одобряю)")
plt.xlabel("Дата")
plt.ylabel("Процент (%)")
plt.grid(True)
plt.show()
```

# **визуализация данных о разногласиях на посту Борисa Ельцинa**

```python
plt.figure(figsize=(18, 8))
plt.plot(df["Дата"], df["Не одобрять"], label="Не одобрять (%)", color="red", marker="o")
plt.title("Процентное изменение поддержки Бориса Ельцина (1994-2000)(Не одобрять)")
plt.xlabel("Дата")
plt.ylabel("Процент (%)")
plt.grid(True)
plt.show()
```

# **композитная визуализация**

```python
plt.figure(figsize=(18, 8))
plt.plot(df["Дата"], df["Не одобрять"], label="Не одобрять (%)", color="red", marker="o")
plt.plot(df["Дата"], df["Одобряю"], label="Одобряю(%)", color="blue", marker="o")
plt.title("Процентное изменение поддержки Бориса Ельцина (1994-2000)")
plt.xlabel("Дата")
plt.ylabel("Процент (%)")
plt.grid(True)
plt.show()
```
